```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['number'].transform(np.max)
```

```python
ords_prods_merge[['user_id', 'number', 'max_order']]
```

```python
ords_prods_merge.loc[ords_prods_merge ['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] > 10) & (ords_prods_merge['max_order'] <= 20) , 'loyalty_flag'] = 'Regular customer'
```

ords_prods_merge.loc[(ords_prods_merge['max_order'] > 10) & (ords_prods_merge['max_order'] <= 20) , 'loyalty_flag'] = 'Regular customer

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts()
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
ords_prods_merge['aver_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id','aver_price']]
```

```python
ords_prods_merge.loc[ords_prods_merge['aver_price']< 10, 'spending_flag'] = 'low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['aver_price']>= 10, 'spending_flag'] = 'high spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge['median_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] > 20, 'freq_customer'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_prior_order'] > 10) & (ords_prods_merge['median_prior_order'] <= 20) , 'freq_customer'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_order'] <= 10, 'freq_customer'] = 'Frequent customer'
```

```python
ords_prods_merge['freq_customer'].value_counts(dropna = False)
```
