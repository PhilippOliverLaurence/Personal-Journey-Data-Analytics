# Merging dfs 

### Import all cleaned df

```python
df_custs = pd.read_csv(os.path.join(path, 'customers.csv'), index_col = False)
```

```python
df_custs
```

```python
df_ord_prod_prior = pd.read_csv(os.path.join(path, 'order_products_prior.csv'), index_col = False)
```

```python
df_ord_prod_prior
```

```python
df_ord = pd.read_csv(os.path.join(path, 'orders.csv'), index_col = False)
```

```python
df_ord
```

```python
df_prod = pd.read_csv(os.path.join(path, 'products.csv'), index_col = False)
```

```python
df_prod
```

```python
df_ord_prior=pd.merge(df_ord_prod_prior,df_ord, on = ['order_id'], indicator = True)
```

```python
df_ord_prior.head()
```

```python
df_ord_prior['_merge'].value_counts(dropna=False)
```

```python
df_ord_prior = df_ord_prior.drop(columns=['_merge'])
```

```python
df_ord_prior.head()
```

```python
df_ord_prod_prior=pd.merge(df_ord_prior,df_prod, on = ['product_id'], indicator = True)
```

```python
df_ord_prod_prior.head()
```

```python
df_ord_prod_prior['_merge'].value_counts(dropna=False)
```

```python
df_ords_prod = df_ord_prod_prior.drop(columns=['_merge'])
```

```python
df=pd.merge(df_ords_prod,df_custs, on = ['user_id'], indicator = True)
```

```python
df.head()
```

```python
df['_merge'].value_counts(dropna=False)
```

```python
df = df.drop(columns=['_merge'])
```
